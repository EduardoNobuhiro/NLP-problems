```python
count = 0

def turnIOB(tokens,label,n):
  IOBLabel=[]
  auxvalue=0
  counter=0
  while counter < len(tokens[n]) :

    if (tokens[n][counter]=='[CLS]' or tokens[n][counter]=='[SEP]' or tokens[n][counter]=='[PAD]') :
      IOBLabel.append('O')
    else:
      if tokens[counter].startswith("##"):
        auxvalue = auxvalue + len(tokens[n][counter])-2
      else:
        auxvalue = auxvalue + len(tokens[n][counter]) + 1
        if (tokens[n][counter - 1] == '[CLS]' or tokens[n][counter - 1] == '[SEP]' or tokens[n][counter - 1] == '[PAD]'):
          auxvalue = auxvalue - 1
    counter1 = 0 
    while counter1 < len(label[n]):
      if label[n][counter1][0] <= auxvalue <= label[n][counter1][1]:
        IOBLabel.append(label[n][counter1][2])
      else:
        IOBLabel.append('O')
    counter+= 1

  return IOBLabel
```
